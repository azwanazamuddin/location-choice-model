```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler

# Load the data
file_path = '../data/Data_MNL_siang.xlsx'
data = pd.read_excel(file_path)

# Data preprocessing
# Dropping non-predictive columns and rows with missing values
data = data.drop(columns=['driver_id', 'trip_jenis_pesanan_Ride', 'trip_jenis_pesanan_Kurir Barang'])
data.dropna(inplace=True)

# Feature scaling
features = data.select_dtypes(include=[np.number]).columns.drop('location_code')
scaler = StandardScaler()
data[features] = scaler.fit_transform(data[features])

# Splitting the data into training and testing sets
X = data.drop('location_code', axis=1)
y = data['location_code']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
rf_classifier.fit(X_train, y_train)

# Predict on the test set
y_pred = rf_classifier.predict(X_test)

# Evaluate the model
print(classification_report(y_test, y_pred))
```
