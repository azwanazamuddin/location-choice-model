```python
import pandas as pd
```

### Load & join all dataset

```python
# Load all kelurahan
all_ = pd.read_excel('../data/agg/kelurahan_bandung.xlsx')
all_ = all_[['OBJECTID', 'kelurahan']]
all_
```

```python
# Load jumlah pasar per kelurahan
pasar = pd.read_excel('../data/agg/pasar_kelurahan.xlsx')
pasar = pasar[['OBJECTID', 'kelurahan', 'jumlah_pasar']]
print(pasar.shape)
pasar.head()
```

```python
# Load jumlah sd  per kelurahan
pend_sd = pd.read_excel('../data/agg/pendidikan_sd_kelurahan.xlsx')
pend_sd = pend_sd[['OBJECTID', 'kelurahan', 'jumlah_sd']]
print(pend_sd.shape)
pend_sd.head()
```

```python
# Load jumlah smp  per kelurahan
pend_smp = pd.read_excel('../data/agg/pendidikan_smp_kelurahan.xlsx')
pend_smp = pend_smp[['OBJECTID', 'kelurahan', 'jumlah_smp']]
print(pend_smp.shape)
pend_smp.head()
```

```python
# Load jumlah sma  per kelurahan
pend_sma = pd.read_excel('../data/agg/pendidikan_sma_kelurahan.xlsx')
pend_sma = pend_sma[['OBJECTID', 'kelurahan', 'jumlah_sma']]
print(pend_sma.shape)
pend_sma.head()
```

```python
# Load jumlah pt per kelurahan
pend_pt = pd.read_excel('../data/agg/pendidikan_pt_kelurahan.xlsx')
pend_pt = pend_pt[['OBJECTID', 'kelurahan', 'jumlah_pt']]
print(pend_pt.shape)
pend_pt.head()
```

```python
# Load jumlah halte per kelurahan
halte = pd.read_excel('../data/agg/transportasi_halte_kelurahan.xlsx')
halte = halte[['OBJECTID', 'kelurahan', 'jumlah_halte']]
print(halte.shape)
halte.head()
```

```python
# Load jumlah stasiun per kelurahan
stasiun = pd.read_excel('../data/agg/transportasi_stasiun_kelurahan.xlsx')
stasiun = stasiun[['OBJECTID', 'kelurahan', 'jumlah_stasiun']]
print(stasiun.shape)
stasiun.head()
```

```python
# Load jumlah terminal per kelurahan
terminal = pd.read_excel('../data/agg/transportasi_terminal_kelurahan.xlsx')
terminal = terminal[['OBJECTID', 'kelurahan', 'jumlah_terminal']]
print(terminal.shape)
terminal.head()
```

```python
# Load jumlah hotel per kelurahan
hotel = pd.read_excel('../data/agg/hotel_kelurahan.xlsx')
hotel = hotel[['OBJECTID', 'kelurahan', 'jumlah_hotel']]
print(hotel.shape)
hotel.head()
```

```python
# Load jumlah mall per kelurahan
mall = pd.read_excel('../data/agg/mall_kelurahan.xlsx')
mall = mall[['OBJECTID', 'kelurahan', 'jumlah_mall']]
print(mall.shape)
mall.head()
```

```python
# Load jumlah ofm per kelurahan
ofm = pd.read_excel('../data/agg/ofm_kelurahan.xlsx')
ofm = ofm[['OBJECTID', 'kelurahan', 'jumlah_ofm']]
print(ofm.shape)
ofm.head()
```

```python
# Load luas guna lahan bangunan permukiman kota per kelurahan
permukiman = pd.read_excel('../data/agg/permukiman_kelurahan.xlsx')
permukiman = permukiman[['OBJECTID', 'kelurahan', 'luas_permukiman_m2']]
print(permukiman.shape)
permukiman.head()
```

```python
# Load luas guna lahan bangunan industri perdagangan dan perkantoran per kelurahan
ind_dagang_kantor = pd.read_excel('../data/agg/industri_dagang_kantor_kelurahan.xlsx')
ind_dagang_kantor = ind_dagang_kantor[['OBJECTID', 'kelurahan', 'luas_ind_m2']]
print(ind_dagang_kantor.shape)
ind_dagang_kantor.head()
```

```python
# Load luas guna lahan hutan, jalur hijau, dan taman kota per kelurahan
taman = pd.read_excel('../data/agg/hutan_jalurhijau_tamankota_kelurahan.xlsx')
taman = taman[['OBJECTID', 'kelurahan', 'luas_taman_m2']]
print(taman.shape)
taman.head()
```

#### Join all dataset

```python
merge_df = pd.merge(all_, pasar, on='OBJECTID', how='left')
merge_df = pd.merge(merge_df, pend_sd, on='OBJECTID', how='left')
merge_df = pd.merge(merge_df, pend_smp, on='OBJECTID', how='left')
merge_df = pd.merge(merge_df, pend_sma, on='OBJECTID', how='left')
merge_df = pd.merge(merge_df, pend_pt, on='OBJECTID', how='left')
merge_df = pd.merge(merge_df, halte, on='OBJECTID', how='left')
merge_df = pd.merge(merge_df, stasiun, on='OBJECTID', how='left')
merge_df = pd.merge(merge_df, terminal, on='OBJECTID', how='left')
merge_df = pd.merge(merge_df, hotel, on='OBJECTID', how='left')
merge_df = pd.merge(merge_df, mall, on='OBJECTID', how='left')
merge_df = pd.merge(merge_df, ofm, on='OBJECTID', how='left')
merge_df = pd.merge(merge_df, permukiman, on='OBJECTID', how='left')
merge_df = pd.merge(merge_df, ind_dagang_kantor, on='OBJECTID', how='left')
final_df = pd.merge(merge_df, taman, on='OBJECTID', how='left')
```

```python
final_df.columns
```

```python
agg_kelurahan = final_df[[
    'OBJECTID',
    'kelurahan',
    'jumlah_pasar',
    'jumlah_sd',
    'jumlah_smp',
    'jumlah_sma',
    'jumlah_pt',
    'jumlah_halte',
    'jumlah_stasiun',
    'jumlah_terminal',
    'jumlah_hotel',
    'jumlah_mall',
    'jumlah_ofm',
    'luas_permukiman_m2',
    'luas_ind_m2',
    'luas_taman_m2'
]]
```

```python
agg_kelurahan.fillna(0, inplace=True)
agg_kelurahan
```

```python
# Load pickup dropoff per kelurahan
pick_drop = pd.read_excel('../data/agg/pickup_dropoff_kelurahan.xlsx')
pick_drop = pick_drop[['OBJECTID', 'kelurahan', 'jumlah_pickup', 'jumlah_dropoff']]
print(pick_drop.shape)
pick_drop.head()
```

```python
add_pick_drop = pd.merge(agg_kelurahan, pick_drop, on='OBJECTID', how='left')
```

```python
# Drop the 'City' column
add_pick_drop.drop(columns='kelurahan_y', inplace=True)
add_pick_drop.rename(columns={'kelurahan_x':'kelurahan'}, inplace=True)
```
